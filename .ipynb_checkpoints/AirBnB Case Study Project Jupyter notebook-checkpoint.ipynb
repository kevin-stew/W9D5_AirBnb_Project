```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head(10)
```

```python
# How many neighborhood groups are available and which shows up the most?

print(air_bnb['neighbourhood_group'].value_counts())
print('-'*40)
print(""" ANSWER: there are 5 Boroughs (aka 'neighborhood_groups') in NYC.
        Manhattan has the most AirBNB listings.""")
```

```python
# Are private rooms the most popular in manhattan?

print(air_bnb[air_bnb['neighbourhood_group'] == 'Manhattan'].groupby('room_type').count()['id'])
print('-'*40)
print(" ANSWER: No. 'Entire home/apt' are the most popular rental 'room_type' in Manhattan")
```

```python
# Which hosts are the busiest and based on their reviews?
print(" ANSWER: Below are the top 10 busiest hosts based on their 'reviews_per_month' value")
air_bnb[['host_id','host_name','reviews_per_month']].sort_values('reviews_per_month', ascending=False).drop_duplicates('host_id').head(10).reset_index(drop=True)
```

```python
#Which neighorhood group has the highest average price?
print("ANSWER: No surprises here, 'Mo-money Manhattan' is the riztiest place to AirBnB whilst in the Big Apple")
air_bnb[['neighbourhood_group','price']].groupby('neighbourhood_group').mean('price').round(1).sort_values('price', ascending=False)
```

```python
# Which neighbor hood group has the highest total price?
print("ANSWER: Brooklyn, Manhattan, and Queens all share the same highest total price of: $10,000ðŸ¤‘")
air_bnb[['neighbourhood_group','price']].groupby('neighbourhood_group').max()
```

```python
#Which top 5 hosts have the highest total price?

print("ANSWER: The most popular host, 'Row NYC', has the most expencieve total price of 100")
air_bnb[['host_id','host_name','reviews_per_month','price']].sort_values('reviews_per_month', ascending=False).drop_duplicates('host_id').head(5).reset_index(drop=True)

```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
print("158* rentals with a review of 100+ currently have zero availability")
print("*(based on row number at bottom of table)")
air_bnb[air_bnb['number_of_reviews'] > 100].loc[air_bnb['availability_365'] == 0]
    
```

```python
# What host has the highest total of prices and where are they located?
print("""ANSWER: Not sure if this is what the question wanted, but below are the top 20 most expensive rentals in the city:

- Kathin of Astoria, 
- Erin of Greenpoint, and 
- Jelena from the Upper West Side 

all tie for the most expensive rental at $10,000.""")
air_bnb[['host_id','host_name','price', 'neighbourhood', 'neighbourhood_group']].sort_values('price', ascending=False).head(20).reset_index(drop=True)
```

```python
# When did Danielle from Queens last receive a review?
print("ANSWER: which 'Danielle form Queens'?")
danielles = air_bnb[air_bnb['host_name'] == 'Danielle']
danielles_from_queens = danielles[air_bnb['neighbourhood_group'] == 'Queens']
danielles_from_queens[['host_id','host_name','neighbourhood_group', 'last_review']].drop_duplicates('host_id').reset_index(drop=True)
```
